```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('movies_sentiment_data.csv')
df.head()
```

```python
# positive 1 and other 0
df['new_sentiment'] = df['sentiment'].apply(lambda x:1 if x == 'positive' else 0)
```

```python
df.head()
```

```python
df['sentiment'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['review'], df['new_sentiment'],test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
X_train_cv = cv.fit_transform(X_train)
X_test_cv = cv.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
model.fit(X_train_cv, y_train)
```

```python
model.score(X_test_cv, y_test)
```

# with removing stop words

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
def remove_stop_words(text):
    doc = nlp(text)
    no_stop_words = [token.text for token in doc if not token.is_stop]
    return ' '.join(no_stop_words)
```

```python
df['review'] = df['review'].apply(remove_stop_words)
```
